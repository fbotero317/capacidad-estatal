# South America

```python
def create_population_density_plots(gdf, continent, output_dir='density_plots'):
    """
    Create population density plots for each country using appropriate ESRI projections.
    
    Parameters:
    -----------
    gdf : GeoDataFrame
        Must contain columns: osm_admin_level, name, population, geometry
    continent : str
        Either 'south_america' or 'africa' to determine the appropriate projection
    output_dir : str
        Directory where plots will be saved
    """
    # Create output directory if it doesn't exist
    Path(output_dir).mkdir(parents=True, exist_ok=True)
    
    # Set the appropriate ESRI projection based on continent
    if continent.lower() == 'south_america':
        # South America Albers Equal Area Conic
        proj_crs = 'esri:102033'
    elif continent.lower() == 'africa':
        # Africa Albers Equal Area Conic
        proj_crs = 'esri:102022'
    else:
        raise ValueError("Continent must be either 'south_america' or 'africa'")
    
    # Project the data
    gdf_projected = gdf.to_crs(proj_crs)
    
    print("The data has been projected to", gdf_projected.crs.name)
    print("The columns pre filter are", gdf_projected.columns, f"dim: {gdf_projected.shape}")

    gdf_filtered = gdf_projected[gdf_projected['osm_admin_level'] == "4"].copy()
    print("The columns filter are", gdf_filtered.columns, f"dim: {gdf_filtered.shape}")

    
    # Calculate area in square kilometers and population density
    gdf_filtered['area_km2'] = gdf_filtered.geometry.area / 10**6  # Convert m² to km²
    gdf_filtered['pop_density'] = gdf_filtered['population'] / gdf_filtered['area_km2']
    print(f"The columns filter and popdensity are {gdf_filtered.columns} and dim {gdf_filtered.shape}")
    # Calculate quintiles for each country
    print("Pre loop")
    print("Iterating on ", gdf_filtered['hasc_country'].unique())
    for country in gdf_filtered['hasc_country'].unique():
        print("Calculating quintiles and density for", country)
        country_data = gdf_filtered[gdf_filtered['hasc_country'] == country]
        
        # Skip if no valid population density values
        if country_data['pop_density'].isna().all():
            print(f"Skipping {country} due to missing population density values")
            continue
            
        # Calculate quintiles for this country
        try:
            quintiles = pd.qcut(country_data['pop_density'], 
                              q=5, 
                              labels=['Q1 (Lowest)', 'Q2', 'Q3', 'Q4', 'Q5 (Highest)'])
            
            country_data.loc[country_data.index, 'density_quintile'] = quintiles
            print(f"Creating quintiles column {country_data.columns}")
            
            # Create the plot
            fig, ax = plt.subplots(1, 1, figsize=(15, 15))
            
            # Plot the country data
            country_data.plot(column='density_quintile',
                            categorical=True,
                            legend=True,
                            ax=ax,
                            cmap='YlOrRd',
                            legend_kwds={'title': 'Population Density\nQuintiles',
                                       'bbox_to_anchor': (1.3, 1)})
            # Plot the country data
            country_data.plot(column='density_quintile',
                              categorical=True,
                              legend=True,
                              ax=ax,
                              cmap='YlOrRd',
                              legend_kwds={'title': 'Population Density\nQuintiles',
                                           'bbox_to_anchor': (1.3, 1),
                                           'frameon': False,  # Removes the legend box outline
                                            'fontsize': 12,    # Increases font size of legend labels
                                            'title_fontsize': 14})  # Increases font size of legend title
            # Calculate total population
            total_pop = country_data['population'].sum()
            
            # Add title and caption
            plt.title(f'Population Density by Administrative Division\n{country}', 
                     pad=20, 
                     size=16)
            
            plt.figtext(0.5, 0.02, 
                       f'Total Population: {total_pop:,.0f}\nProjection: {proj_crs}',
                       ha='center',
                       bbox=dict(facecolor='white', alpha=0.8))
            
            # Remove axes
            ax.axis('off')
            
            # Adjust layout to prevent text cutoff
            plt.tight_layout()
            
            # Save the plot
            print("Saving plot at ", f'{output_dir}/{country.replace(" ", "_")}_density.png')
            plt.savefig(f'{output_dir}/{country.replace(" ", "_")}_density.png', 
                       bbox_inches='tight',
                       dpi=300)
            plt.close()
            
        except ValueError as e:
            print(f"Could not create quintiles for {country}: {str(e)}")
            continue
    
    return gdf_filtered
```

```python

# Example usage for South America
southamerica_density = create_population_density_plots(
    southamerica_kontur, 
    continent='south_america', 
    output_dir=f'{output_path}/southamerica_density_plots'
)

```
